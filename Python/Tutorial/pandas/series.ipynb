```python
fruit_protein = {
    "Avocado": 2.0,
    "Guava": 2.6,
    "Blackberries": 2.0,
    "Oranges": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomogranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}
```

```python
s2 = pd.Series(fruit_protein, name="Protien")
s2
```

```python
# conditional selection:

s2[s2>1]
```

```python
# logical operators
s2[(s2>0.5) & (s2<2)]
```

```python
# not operation
s2[~(s2>1)]
```

```python
# modifying the series:
s2["Mango"] = 2.8
s2
```
