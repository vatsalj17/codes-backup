## creating array from scratch

```python
zeros = np.zeros((3, 4))
print("zeros array: \n", zeros)

ones = np.ones((2, 3))
print("ones array: \n", ones)

full = np.full((2, 4), 9)
print("full array: \n", full)

random = np.random.random((2, 3))
print("random array: \n", random)

sequence = np.arange(0, 10, 2) # 0 to 10 with gap of 2
print("sequence array: \n", sequence)
```

## Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("Matrix: \n", matrix)

tensor = np.array([[[1, 2], [3, 4]], 
                   [[5, 6], [7, 8]]])
print("Tensor:\n", tensor)
```

## array properties

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Shape:", arr.shape)
print("Dimension:", arr.ndim)
print("Size:", arr.size)
print("DType:", arr.dtype)
```

## array reshaping

```python
arr = np.arange(12)
print("original array:", arr)
reshaped = arr.reshape((3, 4))
print("Print reshaped:\n", reshaped)
flattend = reshaped.flatten()
print("Flattened array:", flattend)
ravaled = reshaped.ravel() # returns view, instead of copy
print("raveled array:", ravaled)
transpose = reshaped.T
print("transpose array:\n", transpose)
```
